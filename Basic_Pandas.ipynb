# Lineaer Regression

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()

my_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
my_df.head()
```

```python
my_df['target'] = diabetes.target
```

```python
my_df.head()
```

```python
# Split the data into X and y
X = my_df.drop('target', axis=1)
X.head(1)
```

```python
y = my_df['target']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

# Build a Regression Model

```python
# Just as my previous learning projects
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
 

# Create a Linear Regression Model
lr = LinearRegression()
pf = PolynomialFeatures(degree=1)
model = Pipeline([('Polynomial Features', pf), ('Linear Regression', lr)])
```

```python
# Train the constructed model
model.fit(X_train, y_train)
```

```python
# Perform predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate Performance of the model
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

R2 = r2_score(y_true=y_test, y_pred=y_pred)  # Higher the better: How well the model fits the data.
MSE = mean_squared_error(y_test, y_pred) 
MAE = mean_absolute_error(y_test, y_pred)

print("R2 Score: ", R2)
print("MSE: " ,MSE)
print("MAE: " ,MAE)
```

```python
model.score(X_test, y_test)  # R2 Score
```

```python
model.steps[1][1].coef_
```

```python
model.steps[1][1].rank_
```

```python
model.steps[1][1].intercept_
```

## Plot predicted vs. actual values 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test, y_pred, alpha=0.55)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='orange')
plt.xlabel("Actual Values")
plt.ylabel("Preidcted Values")
plt.title('Perdicted vs. Actual Values (Linear Regression Model)')
```

```python
# Plot the residualas

plt.scatter(y_pred, y_test - y_pred, alpha=0.4)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot (Regression model)")

# If the model is a good fit for the data, the risiduals should show a random scatter of the point around zero with no discernable trend
```
